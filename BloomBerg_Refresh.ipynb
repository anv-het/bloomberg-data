```python
import os,time,subprocess,sys
import xlwings as xw
from pathlib import Path
from xlwings.constants import PasteType
def wait_close_file(xlsx_path :str,values_only_out :str,values_only_out_1 :str,wait_seconds : int =10):
    os.startfile(xlsx_path)
    time.sleep(wait_seconds)
    try:
        app = xw.apps.active
        if app is None:
            raise RuntimeError('No File Found')
        book = None
        for b in app.books:
            try:
                book = b
            except:
                raise RuntimeError('Workbook not found')

        if values_only_out:
            print(f"Creating values-only copy: {values_only_out}")
            vals_book = app.books.add()
                # Copy each sheet's used range as values
            for sh in book.sheets:
                
                # Ensure worksheet exists in target book
                try:
                    target = vals_book.sheets[sh.name]
                except Exception:
                    target = vals_book.sheets.add(sh.name)
                data = sh.used_range.value
                target.range("A1").value = data
                #used = sh.used_range
                #used.api.Copy()
                #target.api.Range("A1").PasteSpecial(Paste=PasteType.xlPasteValuesAndNumberFormats)
                # Remove default empty sheet if present
                try:
                    vals_book.sheets["Sheet1"].delete()
                except Exception:
                    pass

            vals_book.save(values_only_out)
            vals_book.close()

            print("Done.")
            # Save to a new file
            book.save(str(values_only_out_1))

            # Close workbook
            book.close()

    finally:
        if app is not None:
            app.quit()

        
    subprocess.run(["taskkill","/IM",'EXCEL.EXE',"/F"])
wait_close_file(r"C:\blp\data\FA1_vwijagme.xlsx",r"C:\blp\data\FA1_vwijagme_value_copy_1.xlsx",r"C:\blp\data\FA1_vwijagme_value_copy.xlsx")
```

```python
from dropbox.exceptions import ApiError, AuthError
# Creating A Bloomberg On DropBox
def create_folder(dbx: dropbox.Dropbox ,path:str) -> None:
    """Create a folder; prints 'already exists' if present"""
    if not path.startswith("/"):
        path = '/' + path
    try:
        dbx.files_create_folder_v2(path,autorename=False)
        print(f'Created:{path}')
    except ApiError as e:
        if e.error.is_path() and getattr(e.error.get_path(),"is_conflict",lambda:False)():
            print(f'Already exists :{path}')
        else:
            raise
create_folder(dbx,'Bloomberg')
                                        
                                        
```

```python
from pathlib import Path
from dropbox.files import WriteMode, UploadSessionCursor, CommitInfo
from dropbox.exceptions import ApiError, RateLimitError
CHUNK_SIZE = 8 * 10 * 1024 *1024
# --- Ensure folder tree exists ---
def ensure_folder_tree(dbx: dropbox.Dropbox, path: str) -> None:
    path = path.replace("\\", "/")
    if not path.startswith("/"):
        path = "/" + path
    parts = [p for p in path.split("/") if p]
    cur = ""
    for p in parts:
        cur = cur + "/" + p
        try:
            dbx.files_get_metadata(cur)
        except ApiError as e:
    # create if missing
            if e.error.is_path() and getattr(e.error.get_path(), "is_not_found", lambda: False)():
                dbx.files_create_folder_v2(cur, autorename=False)
            else:
                raise
# --- Upload (small or large) ---
def upload_file_to_dropbox(
    local_path: str,
    dropbox_folder: str,
    dropbox_filename: str | None = None,
    overwrite: bool = True,
    ):
    global dbx

    src = Path(local_path).expanduser().resolve()
    if not src.is_file():
        raise FileNotFoundError(src)

    # Normalize target path
    dropbox_folder = dropbox_folder if dropbox_folder.startswith("/") else "/" + dropbox_folder
    dropbox_folder = dropbox_folder.rstrip("/") or "/"

    # Ensure folder exists
    ensure_folder_tree(dbx, dropbox_folder)

    # Decide final filename
    name = dropbox_filename or src.name
    dest_path = f"{dropbox_folder}/{name}"

    mode = WriteMode("overwrite") if overwrite else WriteMode("add")
    size = src.stat().st_size

    if size <= CHUNK_SIZE:
        with src.open("rb") as f:
            meta = dbx.files_upload(f.read(), dest_path, mode=mode, mute=True)
    else:
        with src.open("rb") as f:
            start = dbx.files_upload_session_start(f.read(CHUNK_SIZE))
            cursor = UploadSessionCursor(session_id=start.session_id, offset=f.tell())
            commit = CommitInfo(path=dest_path, mode=mode, mute=True)

            while cursor.offset < size - CHUNK_SIZE:
                dbx.files_upload_session_append_v2(f.read(CHUNK_SIZE), cursor)
                cursor.offset = f.tell()

        dbx.files_upload_session_finish(f.read(size - cursor.offset), cursor, commit)
        meta = dbx.files_get_metadata(dest_path) # optional

        print(f"Uploaded → {meta.path_display} ({size/1024:.1f} KB)")
        return meta

if __name__ == "__main__":
    # Example
    # set env: DROPBOX_APP_KEY, DROPBOX_REFRESH_TOKEN (and optionally DROPBOX_APP_SECRET)
    upload_file_to_dropbox(
    local_path=r"C:\blp\data\FA1_vwijagme_value_copy_1.xlsx",
    dropbox_folder="bloomberg",
    dropbox_filename=None, # or "daily_report.xlsx"
    overwrite=True,
    )
```
